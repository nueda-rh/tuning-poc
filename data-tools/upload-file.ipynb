```python
import os
import boto3
from botocore.exceptions import ClientError

def upload_directory(s3_client, local_directory, bucket, prefix=""):
    for root, dirs, files in os.walk(local_directory):
        for filename in files:
            local_path = os.path.join(root, filename)
            relative_path = os.path.relpath(local_path, local_directory)
            s3_path = os.path.join(prefix, relative_path)
            
            try:
                s3_client.upload_file(local_path, bucket, s3_path)
                print(f"Uploaded {local_path} to {s3_path}")
            except ClientError as e:
                print(f"Error uploading {local_path}: {e}")

# MinIO
s3 = boto3.client(
    "s3",
    endpoint_url=os.getenv('AWS_S3_ENDPOINT', 'http://minio.ic-shared-minio.svc.cluster.local:9000/'),
    aws_access_key_id=os.getenv('AWS_ACCESS_KEY_ID', 'foo'),
    aws_secret_access_key=os.getenv('AWS_SECRET_ACCESS_KEY', 'baa')
)

# MinIO
local_directory = "/opt/app-root/src/.cache/sparsezoo/neuralmagic/mistral-7b-ultrachat200k_mistral_pretrain-pruned50_quantized/"
bucket_name = "models"

# 
upload_directory(s3, local_directory, bucket_name)
```
